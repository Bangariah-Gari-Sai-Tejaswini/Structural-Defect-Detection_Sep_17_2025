## Randomness in the generated text

```python
genai.configure(api_key = os.getenv('GOOGLE_API_KEY'))
model = genai.GenerativeModel('gemini-2.5-flash-lite')
prompt = 'Tell me a joke on engineering'
# Generate response with temperature control
response = model.generate_content(
    prompt,
    generation_config={"temperature": 0}
)

# Print output
print(response.text)
```

METHODS TO CONTROL RANDOMNESS IN THE OUTPUT :

Temperature (Controls the randomness and creativity)
* 0 -> deterministic, focused and factual
* 1 -> balanced creativity
* > 1 -> more diverse and less coherent
* HIgher temperatures may lead to hallucination
